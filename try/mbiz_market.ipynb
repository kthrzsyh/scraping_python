```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime

ItemProductName   = []
ItemProductPrice  = []
ItemSellerName    = []
ItemSellerCity    = []
time              = []
firstCategory     = []
secondCategory    = []

sub_categories = [
    'meetings-incentives-conferences-dan-exhibitions',
    'komputer-dan-perangkat-keras', 
    'elektronik-rumah-tangga', 
    'elektronik-konsumen'
    ]

for sub_category in sub_categories :
  for a in range(1,11) :
    html = urlopen(f"https://www.mbizmarket.co.id/catalog/{sub_category}?page={a}").read()
    soup = BeautifulSoup(html, "lxml")
    print(f"https://www.mbizmarket.co.id/catalog/{sub_category}?page={a}")
    items = soup.find_all("a", "sc-e6026730-10 fAxfCo cardLink")
    len(items)
    for i in items:
      PName   = i.find("h2", "MuiTypography-root sc-e6026730-2 kdlZmg MuiTypography-h2").get_text()
      PPrice  = i.find("span", "sc-e6026730-5 fkDftV").get_text()
      SName   = i.find("span", "MuiTypography-root sc-e6026730-15 bufWrl MuiTypography-caption MuiTypography-colorTextSecondary MuiTypography-displayBlock").get_text()
      SCity   = i.find("span", "sc-e6026730-9 cnetAG").get_text()

      now       = datetime.now()
      ms        = now.microsecond
      timestamp = now.strftime("%Y-%m-%d %H:%M:%S") + f".{ms:03d}"

      if PPrice != "":
        ItemProductName.append(PName)
        ItemProductPrice.append(PPrice)
        ItemSellerName.append(SName)
        ItemSellerCity.append(SCity)
        firstCategory.append('Elektronik, Komputer & Komunikasi')
        secondCategory.append(sub_category)
        time.append(timestamp)

      dict_items = {
          "ProductName": ItemProductName, 
          "ProductPrice":ItemProductPrice, 
          "SellerName":ItemSellerName, 
          "SellerCity" : ItemSellerCity,
          "First Category" : firstCategory, 
          "Second Category" : secondCategory, 
          "Timestamp" : time}
      df = pd.DataFrame(dict_items, columns = [
          "ProductName", 
          "ProductPrice", 
          "SellerName", 
          "SellerCity",
          "First Category",
          "Second Category",
          "Timestamp"])
df  
  
```
